```python
from pyecharts import Map
```

```python
map=Map("各省白酒市场容量", "单位：亿元", 
        title_color="#fff", 
        title_pos="center", 
        width=1000,  
        height=600, 
        background_color='#404a59')

attr = df['province']
value = df['mkt_volumn']
map.add("", attr, value, 
        visual_range=[min(value), max(value)],
        visual_text_color="#000", 
        symbol_size=15,
        is_visualmap=True,
        is_label_show=True,
        label_pos = 'bottom')
map
```
