```python
import style2
```

```python
path_content = 'scenery/s_verticle.jpg'
path_style = 'scenery/style_flower.jpg'
```

```python
style = style2.StyleTransfer(path_content, path_style)
```

```python
style.show_inputs()
```

```python
style.optimize(iterations=5000, learning_rate=1)
```

```python
style.show_results()
```
