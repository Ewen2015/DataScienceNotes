```python
# 1. states

prod_s = d[d['STATES'] == '在售']
prod_s.shape
```

```python
# 2. chan_no

'手机银行' in d.iloc[0]['CHAN_NO'].split(sep=',')
```

```python
m = d['CHAN_NO'].apply(lambda x: '手机银行' in x.split(sep=','))
```

```python
m
```

```python
prod_m = d[m]
prod_m.shape
```

```python
prod_m.tail()
```

```python
# 3. zone

z = []

for i in prod_m['ZONE']:
    z += i.split(sep=',')
```

```python
import itertools
```

```python
z_df = pd.DataFrame(z, columns=['branch'])
z_df['cnt'] = 1
z_cnt = z_df.groupby(by='branch').count().reset_index().sort_values(by='cnt', ascending=False).reset_index(drop=True)
```

```python
z_cnt
```

```python
z_cnt.to_csv(p+'branchCount.csv', index=False)
```

```python
def groupbyCount(df, col):
    c = df[[col]]
    c['cnt'] = 1
    return c.groupby(col).count().sort_values(by='cnt', ascending=False).reset_index()
```

```python
r_cnt = groupbyCount(prod_s, 'RISK_LEVEL')
r_cnt
```

```python
r_cnt.to_csv(p+'riskCount.csv', index=False)
```

```python
cat_cnt = groupbyCount(prod_s, 'PARAM_NAME')
cat_cnt
```
